```python
import numpy
import pandas 
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
military = pandas.read_csv('Military Expenditure and GDP.csv')
military
```

```python
military = military.replace({',': ''}, regex=True) #Using replace function to remove ',' between numbers to avoid 
#errors while computation
```

```python
military
```

```python
military.loc[:,'ME2010':'ME2015'] = military.loc[:,'ME2010':'ME2015'].astype(int)
#pandas.options.display.float_format = '{:,.5f}'.format
military.loc[:,'%ME2010':'%ME2015'] = military.loc[:,'%ME2010':'%ME2015'].astype(float)
```

```python
# Calculating Total GDP in number from the GDP percentage share in military expenditure
i = 0
for i in range(6):
    military['tot_GDP201'+ str(i)]= military['ME201'+str(i)]/military['%ME201'+str(i)]
    military['tot_GDP201'+str(i)] = military['tot_GDP201'+str(i)].astype(int)
```

```python
%matplotlib inline
f, ax = plt.subplots(6,figsize=(20,35))
for i in range(6):
    bar_width = 0.5
    x= numpy.arange(14)
    ax[i].bar(x,military['tot_GDP201'+str(i)] , bar_width, color="green",label='Total GDP')
    ax[i].bar(x-bar_width,military['ME201'+str(i)] , bar_width, color="red",bottom = military['tot_GDP201'+str(i)],label='Military Expenditure')
    ax[i].set_xticks(x + bar_width/4)
    ax[i].set_xticklabels(military['Country'])
    ax[i].set_title('Comparision of Military expenditure and GDP of each Country in 201'+str(i))
    ax[i].set_ylabel('Expenditure in Million Dollars')
    ax[i].set_xlabel('Countries')
    ax[i].legend()
fig.show()
```

```python
from scipy import stats
x = numpy.arange(6)
slopes = [0]*14
for i in range(14):
    y = numpy.array(military_new.ix[i,1:7])
    slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
    slopes[i] = slope
military_new['Rate_of_Change'] = slopes
```

```python
military_new['Change_in_Percentage_ME'] = ((military_new['Rate_of_Change']*6)/military_new['ME2010'])*100
```

```python
military_new['Change_in_Percentage_ME']
```

```python
military_new['Change_in_Value_ME'] = (military_new['ME2010']* military_new['Change_in_Percentage_ME'])/100
```

```python
military_new['Change_in_Value_ME']
```

```python
Growing_countries = military_new.loc[military_new['Change_in_Value_ME'] > 2000]
```
